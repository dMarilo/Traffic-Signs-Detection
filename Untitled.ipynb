```python
"""
Ovaj blok koda definiše funkciju koja učitava podatke iz pickle fajla 
i vraća samo tražene kolone iz tog fajla kao tuple (nepromjenjiv, uredni skup vrijednosti).
"""

import pickle

def load_pickled_data(file, columns):
    """
    Učitava podatke iz pickle fajla za treniranje i testiranje.
    
    Parametri
    ----------
    file    :   
              Ime pickle fajla.
    columns : list of strings
              Lista kolona iz pickle podataka koje nas interesuju.

    Povratna vrijednost
    -------
    Tuple skupova podataka za zadane kolone.    
    """

    # Otvaranje pickle fajla u binarnom režimu za čitanje
    with open(file, mode='rb') as f:
        # Učitavanje podataka iz pickle fajla
        dataset = pickle.load(f)
    # Vraćanje tuple sa vrijednostima iz dataset-a za tražene kolone
    return tuple(map(lambda c: dataset[c], columns))
```

```python
"""
Ovaj blok koda sadrži pomoćne funkcije za mjerenje proteklog vremena i 
prikaz napretka u terminalu pomoću progres bara.
"""

import sys
import time

def get_time_hhmmss(start = None):
    """
    Računa proteklo vrijeme od start-a i formatira ga kao string.
    Ako nije proslijedjen početak, vraća trenutno vrijeme.
    """
    if start is None:
        # Vraća trenutno vrijeme u formatu YYYY/MM/DD HH:MM:SS
        return time.strftime("%Y/%m/%d %H:%M:%S")
    
    # Vremenska oznaka završetka
    end = time.time()
    # Izračunava proteklo vrijeme u minutama i sekundama
    m, s = divmod(end - start, 60)
    # Izračunava proteklo vrijeme u satima i minutama
    h, m = divmod(m, 60)
    # Formatira vrijeme u obliku hh:mm:ss
    time_str = "%02d:%02d:%02d" % (h, m, s)
    return time_str   

# Ispis progres bara u terminalu
def print_progress(iteration, total):
    """
    Poziva se u petlji da bi se prikazao progres bar u terminalu
    
    Parametri
    ----------
        
    iteration : 
                Trenutna iteracija (int)
    total     : 
                Ukupan broj iteracija (int)
    """
    # Formatiranje postotka sa 0 decimalnih mjesta
    str_format = "{0:.0f}"
    percents = str_format.format(100 * (iteration / float(total)))
    # Dužina ispunjenog dijela progress bara (od 0 do 100)
    filled_length = int(round(100 * iteration / float(total)))
    # Kreiranje stringa progress bara sa ispunjenim i praznim dijelom
    bar = '█' * filled_length + '-' * (100 - filled_length)

    # Ispis progress bara u istom redu (\r vraća kursor na početak linije)
    sys.stdout.write('\r |%s| %s%%' % (bar, percents)),

    # Ako je završena zadnja iteracija, prelazak u novi red
    if iteration == total:
        sys.stdout.write('\n')
    sys.stdout.flush()
```

```python
"""
Ovaj blok koda definiše funkcije za preprocesiranje slike i labela u datasetu:
- skaliranje piksela i konverziju u sivu skalu,
- poboljšanje kontrasta slike,
- one-hot enkodiranje labela,
- miješanje podataka,
kao i učitavanje pickle fajla i vraćanje preprocesiranih podataka.
"""

from sklearn.utils import shuffle
from skimage import exposure
import warnings 

# Broj klasa u datasetu
num_classes = 43

def preprocess_dataset(X, y = None):
    """
    Vrši skaliranje karakteristika, one-hot enkodiranje labela i miješanje podataka ako su labeli dostupni.
    Pretpostavlja da je originalni dataset sortiran po labelama.
    
    Parametri
    ----------
    X                : ndarray
                       Niz koji sadrži primjere sa karakteristikama (npr. slike).
    y                : ndarray, opcionalno, default je `None`
                       Labeli dataset-a u obliku indeksa klasa.
    
    Povratna vrijednost
    -------
    Tuple koji sadrži preprocesirane X i y.    
    """
    print("Preprocesiranje dataset-a sa {} primjera:".format(X.shape[0]))

    # Konverzija RGB slike u sivu skalu koristeći standardne težinske koeficijente za kanale
    X = 0.299 * X[:, :, :, 0] + 0.587 * X[:, :, :, 1] + 0.114 * X[:, :, :, 2]

    # Skaliranje piksela na interval [0, 1]
    X = (X / 255.).astype(np.float32)
        
    # Primjena CLAHE (Contrast Limited Adaptive Histogram Equalization) za poboljšanje kontrasta slike
    for i in range(X.shape[0]):
        # Ignorišemo upozorenja tokom obrade slike
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            X[i] = exposure.equalize_adapthist(X[i])
        # Ispis progres bara tokom procesiranja
        print_progress(i + 1, X.shape[0])
        
    # Ako su dostupni labeli, pretvaramo ih u one-hot kodiranje i miješamo podatke
    if y is not None:  
        # One-hot enkodiranje labela (npr. labela 3 postaje [0, 0, 0, 1, 0, ...])
        y = np.eye(num_classes)[y]
        # Miješanje podataka (x i y) radi boljeg treniranja modela
        X, y = shuffle(X, y)

    # Dodajemo dimenziju kanala za sivu skalu (npr. oblik postaje (n_samples, visina, širina, 1))
    X = X.reshape(X.shape + (1,)) 
    return X, y

def class_name(one_hot):
    # Pretvara one-hot vektor u naziv klase (koristeći indeks jedinice u vektoru)
    return signnames[one_hot.nonzero()[0][0]]

def load_and_process_data(pickled_data_file):
    """
    Učitava pickle-ovane podatke i preprocesira slike i labele skaliranjem, miješanjem i one-hot enkodiranjem.
    
    Parametri
    ----------
    pickled_data_file  : string
                         Ime pickle fajla sa podacima.
    
    Povratna vrijednost
    -------
    Tuple X i y sa preprocesiranim slikama i labelama.
    """
    # Učitavamo sirove podatke sa slikama i labelama
    X, y = load_pickled_data(pickled_data_file, columns = ['features', 'labels'])
    # Preprocesiramo dataset (skaliranje, contrast enhancement, one-hot encoding, shuffle)
    X, y = preprocess_dataset(X, y)
    return (X, y)
```

```python
from nolearn.lasagne import BatchIterator
from skimage.transform import rotate
from skimage.transform import warp
from skimage.transform import ProjectiveTransform
import random

class AugmentedSignsBatchIterator(BatchIterator):
    """
    Iterates over dataset in batches. 
    Allows images augmentation by randomly rotating, applying projection, 
    adjusting gamma, blurring, adding noize and flipping horizontally.
    """
        
    def __init__(self, batch_size, shuffle = False, seed = 42, p = 0.5, intensity = 0.5):
        """
        Initialises an instance with usual iterating settings, as well as data augmentation coverage
        and augmentation intensity.
        
        Parameters
        ----------
        batch_size:
                    Size of the iteration batch.
        shuffle   :
                    Flag indicating if we need to shuffle the data.
        seed      :
                    Random seed.
        p         :
                    Probability of augmenting a single example, should be in a range of [0, 1] .
                    Defines data augmentation coverage.
        intensity :
                    Augmentation intensity, should be in a [0, 1] range.
        
        Returns
        -------
        New batch iterator instance.
        """
        super(AugmentedSignsBatchIterator, self).__init__(batch_size, shuffle, seed)
        self.p = p
        self.intensity = intensity

    def transform(self, Xb, yb):
        """
        Applies a pipeline of randomised transformations for data augmentation.
        """
        Xb, yb = super(AugmentedSignsBatchIterator, self).transform(
            Xb if yb is None else Xb.copy(), 
            yb
        )
        
        if yb is not None:
            batch_size = Xb.shape[0]
            image_size = Xb.shape[1]
            
            Xb = self.rotate(Xb, batch_size)
            Xb = self.apply_projection_transform(Xb, batch_size, image_size)

        return Xb, yb
        
    def rotate(self, Xb, batch_size):
        """
        Applies random rotation in a defined degrees range to a random subset of images. 
        Range itself is subject to scaling depending on augmentation intensity.
        """
        for i in np.random.choice(batch_size, int(batch_size * self.p), replace = False):
            delta = 30. * self.intensity # scale by self.intensity
            Xb[i] = rotate(Xb[i], random.uniform(-delta, delta), mode = 'edge')
        return Xb   
    
    def apply_projection_transform(self, Xb, batch_size, image_size):
        """
        Applies projection transform to a random subset of images. Projection margins are randomised in a range
        depending on the size of the image. Range itself is subject to scaling depending on augmentation intensity.
        """
        d = image_size * 0.3 * self.intensity
        for i in np.random.choice(batch_size, int(batch_size * self.p), replace = False):        
            tl_top = random.uniform(-d, d)     # Top left corner, top margin
            tl_left = random.uniform(-d, d)    # Top left corner, left margin
            bl_bottom = random.uniform(-d, d)  # Bottom left corner, bottom margin
            bl_left = random.uniform(-d, d)    # Bottom left corner, left margin
            tr_top = random.uniform(-d, d)     # Top right corner, top margin
            tr_right = random.uniform(-d, d)   # Top right corner, right margin
            br_bottom = random.uniform(-d, d)  # Bottom right corner, bottom margin
            br_right = random.uniform(-d, d)   # Bottom right corner, right margin

            transform = ProjectiveTransform()
            transform.estimate(np.array((
                    (tl_left, tl_top),
                    (bl_left, image_size - bl_bottom),
                    (image_size - br_right, image_size - br_bottom),
                    (image_size - tr_right, tr_top)
                )), np.array((
                    (0, 0),
                    (0, image_size),
                    (image_size, image_size),
                    (image_size, 0)
                )))
            Xb[i] = warp(Xb[i], transform, output_shape=(image_size, image_size), order = 1, mode = 'edge')

        return Xb
```

```python
"""
Ovaj blok koda sadrži dvije funkcije za augmentaciju dataseta slika:
1. flip_extend proširuje dataset tako što horizontalno i vertikalno okreće slike određenih klasa,
pazeći da slike ostanu u odgovarajućim klasama neke klase se mapiraju na iste, neke na druge klase.
2. extend_balancing_classes proširuje dataset dupliciranjem i augmentacijom slika radi balansiranja klasa,
tako da sve klase imaju približno isti broj primjera, uz kontrolu intenziteta augmentacije.
"""

def flip_extend(X, y):
    """
    Proširuje dataset okretanjem slika određenih klasa horizontalno i/ili vertikalno,
    čime se povećava broj primjera zadržavajući njihove klase ili ih prebacujući u povezane klase.

    Parametri
    ----------
    X : ndarray
        Niz slika sa originalnim primjerima.
    y : ndarray
        Niz labela u obliku indeksa klasa.

    Povratna vrijednost
    -------
    Tuple (X_extended, y_extended) sa proširenim podacima i labelama.
    """
    # Klase čije slike nakon horizontalnog okretanja ostaju iste klase
    self_flippable_horizontally = np.array([11, 12, 13, 15, 17, 18, 22, 26, 30, 35])
    # Klase čije slike nakon vertikalnog okretanja ostaju iste klase
    self_flippable_vertically = np.array([1, 5, 12, 15, 17])
    # Klase čije slike nakon horizontalnog i vertikalnog okretanja ostaju iste klase
    self_flippable_both = np.array([32, 40])
    # Parovi klasa koje se preslikavaju jedna u drugu nakon horizontalnog okretanja
    cross_flippable = np.array([
        [19, 20], 
        [33, 34], 
        [36, 37], 
        [38, 39],
        [20, 19], 
        [34, 33], 
        [37, 36], 
        [39, 38],   
    ])
    num_classes = 43
    
    # Inicijalizacija praznih nizova za proširene podatke i labele
    X_extended = np.empty([0, X.shape[1], X.shape[2], X.shape[3]], dtype = X.dtype)
    y_extended = np.empty([0], dtype = y.dtype)
    
    # Prolaz kroz sve klase radi proširenja podataka
    for c in range(num_classes):
        # Dodajemo originalne slike klase c u prošireni dataset
        X_extended = np.append(X_extended, X[y == c], axis = 0)
        
        # Ako klasa može biti horizontalno okrenuta i ostati ista klasa
        if c in self_flippable_horizontally:
            # Dodajemo horizontalno okrenute slike iste klase
            X_extended = np.append(X_extended, X[y == c][:, :, ::-1, :], axis = 0)
        
        # Ako klasa može biti horizontalno okrenuta i preslikati se u drugu klasu
        if c in cross_flippable[:, 0]:
            # Nalazimo klasu u koju se preslikava
            flip_class = cross_flippable[cross_flippable[:, 0] == c][0][1]
            # Dodajemo horizontalno okrenute slike te povezane klase
            X_extended = np.append(X_extended, X[y == flip_class][:, :, ::-1, :], axis = 0)
        
        # Dodajemo odgovarajuće labele za nove slike koje pripadaju klasi c
        y_extended = np.append(y_extended, np.full((X_extended.shape[0] - y_extended.shape[0]), c, dtype = int))
        
        # Ako klasa može biti vertikalno okrenuta i ostati ista klasa
        if c in self_flippable_vertically:
            # Dodajemo vertikalno okrenute slike iste klase
            X_extended = np.append(X_extended, X_extended[y_extended == c][:, ::-1, :, :], axis = 0)
        # Dodajemo labele za te nove slike
        y_extended = np.append(y_extended, np.full((X_extended.shape[0] - y_extended.shape[0]), c, dtype = int))
        
        # Ako klasa može biti okrenuta i horizontalno i vertikalno i ostati ista klasa
        if c in self_flippable_both:
            # Dodajemo slike okrenute i horizontalno i vertikalno
            X_extended = np.append(X_extended, X_extended[y_extended == c][:, ::-1, ::-1, :], axis = 0)
        # Dodajemo labele za te nove slike
        y_extended = np.append(y_extended, np.full((X_extended.shape[0] - y_extended.shape[0]), c, dtype = int))
    
    # Vraćamo proširene podatke i labele
    return (X_extended, y_extended)

def extend_balancing_classes(X, y, aug_intensity = 0.5, counts = None):
    """
    Proširuje dataset dupliciranjem i augmentacijom slika kako bi se balansirao broj primjera po klasama.
    Intenzitet augmentacije i ciljane veličine klasa mogu se podesiti.

    Parametri
    ----------
    X : ndarray
        Niz slika sa originalnim primjerima.
    y : ndarray
        Niz labela u obliku indeksa klasa.
    aug_intensity : float
        Intenzitet augmentacije u rasponu [0, 1].
    counts : lista ili None
        Ciljni broj primjera po klasi. Ako None, koristi se maksimalan broj primjera iz originalnih klasa.

    Povratna vrijednost
    -------
    Tuple (X_extended, y_extended) sa proširenim i augmentiranim podacima i labelama.
    """
    num_classes = 43

    # Izračunava broj primjera po klasama
    _, class_counts = np.unique(y, return_counts = True)
    max_c = max(class_counts)
    total = max_c * num_classes if counts is None else np.sum(counts)
    
    # Inicijalizacija praznih nizova za proširene podatke i labele
    X_extended = np.empty([0, X.shape[1], X.shape[2], X.shape[3]], dtype = np.float32)
    y_extended = np.empty([0], dtype = y.dtype)
    print("Extending dataset using augmented data (intensity = {}):".format(aug_intensity))
    
    # Petlja kroz klase radi proširenja i augmentacije podataka
    for c, c_count in zip(range(num_classes), class_counts):
        # Određuje ciljanu veličinu klase (ciljani broj primjera)
        max_c = max_c if counts is None else counts[c]
        
        # Uzima originalne slike klase i normalizuje ih na [0,1]
        X_source = (X[y == c] / 255.).astype(np.float32)
        y_source = y[y == c]
        
        # Dodaje originalne slike u prošireni dataset
        X_extended = np.append(X_extended, X_source, axis = 0)
        
        # Ponavlja augmentaciju i dodavanje slika onoliko puta koliko je potrebno da se dostigne cilj
        for i in range((max_c // c_count) - 1):
            batch_iterator = AugmentedSignsBatchIterator(batch_size = X_source.shape[0], p = 1.0, intensity = aug_intensity)
            for x_batch, _ in batch_iterator(X_source, y_source):
                X_extended = np.append(X_extended, x_batch, axis = 0)
                print_progress(X_extended.shape[0], total)
        
        # Dodatna augmentacija za ostatak koji nije djeljiv sa brojem originalnih primjera
        batch_iterator = AugmentedSignsBatchIterator(batch_size = max_c % c_count, p = 1.0, intensity = aug_intensity)
        for x_batch, _ in batch_iterator(X_source, y_source):
            X_extended = np.append(X_extended, x_batch, axis = 0)
            print_progress(X_extended.shape[0], total)
            break
        
        # Dodaje labele za novo generisane slike klase c
        added = X_extended.shape[0] - y_extended.shape[0]
        y_extended = np.append(y_extended, np.full((added), c, dtype = int))
        
    # Vraća podatke skalirane nazad u opseg [0,255] i njihove labele
    return ((X_extended * 255.).astype(np.uint8), y_extended)
```

```python
from collections import namedtuple

Parameters = namedtuple('Parameters', [
        # Data parameters
        'num_classes', 'image_size', 
        # Training parameters
        'batch_size', 'max_epochs', 'log_epoch', 'print_epoch',
        # Optimisations
        'learning_rate_decay', 'learning_rate',
        'l2_reg_enabled', 'l2_lambda', 
        'early_stopping_enabled', 'early_stopping_patience', 
        'resume_training', 
        # Layers architecture
        'conv1_k', 'conv1_d', 'conv1_p', 
        'conv2_k', 'conv2_d', 'conv2_p', 
        'conv3_k', 'conv3_d', 'conv3_p', 
        'fc4_size', 'fc4_p'
    ])
```

```python
"""
Ova klasa Paths olakšava upravljanje i pristupanje direktorijumima i fajlovima 
koji se koriste za čuvanje modela, istorije treniranja i grafika učenja.
Na osnovu parametara modela kreira se jedinstveni naziv foldera i fajlova,
a ako folder ne postoji, automatski se kreira.
"""

import os

class Paths(object):
    """
    Omogućava jednostavan pristup i upravljanje fajlovima i folderima vezanim za
    čuvanje podataka tokom treniranja modela.
    """
    
    def __init__(self, params):
        """
        Inicijalizuje Paths objekat, kreira foldere ako ne postoje i određuje putanje.
        
        Parameters
        ----------
        params : Parameters
            namedtuple sa parametrima modela, koji se koriste za generisanje naziva fajlova i foldera.
        """
        # Generiše ime modela na osnovu arhitekture i optimizacija
        self.model_name = self.get_model_name(params)
        # Generiše naziv varijabilnog scope-a modela (koristi se u TensorFlow-u)
        self.var_scope = self.get_variables_scope(params)
        # Root folder za čuvanje modela: trenutni radni folder + /models/ + naziv modela
        self.root_path = os.getcwd() + "/models/" + self.model_name + "/"
        # Putanja do fajla modela
        self.model_path = self.get_model_path()
        # Putanja do fajla sa istorijom treniranja
        self.train_history_path = self.get_train_history_path()
        # Putanja do slike sa grafikonima učenja
        self.learning_curves_path = self.get_learning_curves_path()
        # Kreira folder ako ne postoji
        os.makedirs(self.root_path, exist_ok = True)

    def get_model_name(self, params):
        """
        Kreira jedinstveni naziv modela enkodiranjem važnih parametara modela.
        
        Parameters
        ----------
        params : Parameters
            Parametri modela.
        
        Returns
        -------
        string : ime modela
        """
        # Formira string sa arhitektonskim parametrima (kernel size, dubina, padding za konvolucione slojeve, fc veličina)
        model_name = "k{}d{}p{}_k{}d{}p{}_k{}d{}p{}_fc{}p{}".format(
            params.conv1_k, params.conv1_d, params.conv1_p, 
            params.conv2_k, params.conv2_d, params.conv2_p, 
            params.conv3_k, params.conv3_d, params.conv3_p, 
            params.fc4_size, params.fc4_p
        )
        # Dodaje informacije o tome da li je uključeno smanjenje stope učenja
        model_name += "_lrdec" if params.learning_rate_decay else "_no-lrdec"
        # Dodaje informacije o tome da li je uključena L2 regularizacija
        model_name += "_l2" if params.l2_reg_enabled else "_no-l2"
        return model_name

    def get_variables_scope(self, params):
        """
        Kreira naziv scope-a za varijable modela enkodiranjem važnih parametara.
        
        Parameters
        ----------
        params : Parameters
            Parametri modela.
        
        Returns
        -------
        string : naziv scope-a
        """
        var_scope = "k{}d{}_k{}d{}_k{}d{}_fc{}_fc0".format(
            params.conv1_k, params.conv1_d,
            params.conv2_k, params.conv2_d,
            params.conv3_k, params.conv3_d, 
            params.fc4_size
        )
        return var_scope

    def get_model_path(self):
        """
        Vrati punu putanju do fajla modela.
        
        Returns
        -------
        string : putanja do model.ckpt fajla
        """
        return self.root_path + "model.ckpt"

    def get_train_history_path(self):
        """
        Vrati punu putanju do fajla sa istorijom treniranja.
        
        Returns
        -------
        string : putanja do fajla train_history
        """
        return self.root_path + "train_history"
    
    def get_learning_curves_path(self):
        """
        Vrati punu putanju do slike sa grafikonima učenja.
        
        Returns
        -------
        string : putanja do learning_curves.png fajla
        """
        return self.root_path + "learning_curves.png"
```

```python
class EarlyStopping(object):
    """
    Prati metriku i prekida trening ako nema poboljšanja nakon određenog broja epoha.
    Pri tome vraća model na stanje sa najboljom metrikom.
    """
    
    def __init__(self, saver, session, patience=100, minimize=True):
        """
        Params:
        - saver: TensorFlow saver objekat za čuvanje i vraćanje modela.
        - session: TensorFlow sesija u kojoj se model trenira.
        - patience: broj epoha koje čekamo pre nego što zaustavimo trening.
        - minimize: da li je cilj minimizirati metriku (npr. grešku) ili maksimizirati (npr. tačnost).
        """
        self.minimize = minimize
        self.patience = patience
        self.saver = saver
        self.session = session
        # Početna "najbolja" vrednost
        self.best_monitored_value = np.inf if minimize else 0.
        self.best_monitored_epoch = 0
        self.restore_path = None

    def __call__(self, value, epoch):
        """
        Poziva se posle svake epohe sa aktuelnom vrednošću metrike i epohe.
        Vraća True ako je vreme da se zaustavi trening i učita najbolji model.
        """
        # Ako je trenutna metrika bolja od najbolje (u zavisnosti od minimize)
        if (self.minimize and value < self.best_monitored_value) or \
           (not self.minimize and value > self.best_monitored_value):
            self.best_monitored_value = value
            self.best_monitored_epoch = epoch
            # Sačuvaj model na disk
            self.restore_path = self.saver.save(self.session, os.getcwd() + "/early_stopping_checkpoint")
        
        # Provera da li je prošlo više epoha od strpljenja bez poboljšanja
        elif self.best_monitored_epoch + self.patience < epoch:
            if self.restore_path is not None:
                # Vraćanje na najbolji model
                self.saver.restore(self.session, self.restore_path)
            else:
                print("ERROR: Failed to restore session")
            return True  # signal za prekid treninga
        
        return False
```

```python
class SimpleModelLogger:
    def __init__(self, logfile=None):
        """
        Jednostavan logger koji štampa poruke na ekran i opciono ih beleži u fajl.
        
        Parametri
        ----------
        logfile : str ili None
            Ako je prosleđen, poruke će se dodatno upisivati u ovaj fajl.
        """
        self.logfile = logfile

    def log(self, message):
        """
        Štampa poruku na ekran i upisuje je u fajl (ako postoji).
        
        Parametri
        ----------
        message : str
            Tekstualna poruka koja se loguje.
        """
        print(message)
        if self.logfile:
            with open(self.logfile, 'a') as f:
                f.write(message + '\n')

    def log_parameters(self, params, train_size, valid_size, test_size):
        """
        Loguje osnovne informacije o dataset-u i parametrima modela.
        
        Parametri
        ----------
        params     : Parameters
                     Namedtuple koji sadrži parametre modela i treniranja.
        train_size : int
                     Broj primera u trening skupu.
        valid_size : int
                     Broj primera u validacionom skupu.
        test_size  : int
                     Broj primera u test skupu.
        """
        if params.resume_training:
            self.log("=============================================")
            self.log("============= RESUMING TRAINING =============")
            self.log("=============================================")

        self.log("=================== DATA ====================")
        self.log(f"            Training set: {train_size} examples")
        self.log(f"          Validation set: {valid_size} examples")
        self.log(f"             Testing set: {test_size} examples")
        self.log(f"              Batch size: {params.batch_size}")

        self.log("=================== MODEL ===================")
        self.log("--------------- ARCHITECTURE ----------------")
        self.log(f"{'':10} {'Type':10} {'Size':8} {'Dropout (keep p)':15}")
        self.log(f"{'Layer 1':10} {params.conv1_k}x{params.conv1_k} Conv   {str(params.conv1_d):8} {str(params.conv1_p):15}")
        self.log(f"{'Layer 2':10} {params.conv2_k}x{params.conv2_k} Conv   {str(params.conv2_d):8} {str(params.conv2_p):15}")
        self.log(f"{'Layer 3':10} {params.conv3_k}x{params.conv3_k} Conv   {str(params.conv3_d):8} {str(params.conv3_p):15}")
        self.log(f"{'Layer 4':10} FC        {str(params.fc4_size):8} {str(params.fc4_p):15}")
        
        self.log("---------------- PARAMETERS -----------------")
        self.log("     Learning rate decay: " + ("Enabled" if params.learning_rate_decay else f"Disabled (rate = {params.learning_rate})"))
        self.log("       L2 Regularization: " + (f"Enabled (lambda = {params.l2_lambda})" if params.l2_reg_enabled else "Disabled"))
        self.log("          Early stopping: " + (f"Enabled (patience = {params.early_stopping_patience})" if params.early_stopping_enabled else "Disabled"))
        self.log(" Keep training old model: " + ("Enabled" if params.resume_training else "Disabled"))

```

```python
def fully_connected(input, size):
    """
    Izračunava sloj potpuno povezanih neurona (fully connected layer).
    Povratna vrednost je linearna kombinacija: input * weights + bias.
    
    Parameters
    ----------
    input : Tensor
        Ulazni tensor (obično izlaz prethodnog sloja).
    size : int
        Broj neurona u ovom sloju.
    
    Returns
    -------
    Tensor sa izračunatim vrednostima sloja.
    """
    weights = tf.get_variable('weights', 
        shape=[input.get_shape()[1], size],
        initializer=tf.contrib.layers.xavier_initializer()
    )
    biases = tf.get_variable('biases',
        shape=[size],
        initializer=tf.constant_initializer(0.0)
    )
    return tf.matmul(input, weights) + biases


def fully_connected_relu(input, size):
    """
    Fully connected sloj sa ReLU aktivacijom.
    
    Parameters
    ----------
    input : Tensor
        Ulazni tensor.
    size : int
        Broj neurona u sloju.
    
    Returns
    -------
    Tensor sa ReLU aktivacijom.
    """
    return tf.nn.relu(fully_connected(input, size))


def conv_relu(input, kernel_size, depth):
    """
    Konvolucioni sloj sa ReLU aktivacijom.
    
    Parameters
    ----------
    input : Tensor
        Ulazni tensor (npr. slika ili prethodni sloj).
    kernel_size : int
        Dimenzije kernel matrice (npr. 3 za 3x3 kernel).
    depth : int
        Broj filtera (dubina izlaza).
    
    Returns
    -------
    Tensor nakon konvolucije i ReLU aktivacije.
    """
    weights = tf.get_variable('weights', 
        shape=[kernel_size, kernel_size, input.get_shape()[3], depth],
        initializer=tf.contrib.layers.xavier_initializer()
    )
    biases = tf.get_variable('biases',
        shape=[depth],
        initializer=tf.constant_initializer(0.0)
    )
    conv = tf.nn.conv2d(input, weights,
        strides=[1, 1, 1, 1], padding='SAME')
    return tf.nn.relu(conv + biases)


def pool(input, size):
    """
    Max-pooling sloj.
    
    Parameters
    ----------
    input : Tensor
        Ulazni tensor (obično izlaz konvolucionog sloja).
    size : int
        Dimenzija prozora za pooling (npr. 2 za 2x2).
    
    Returns
    -------
    Tensor nakon max pooling operacije.
    """
    return tf.nn.max_pool(
        input,
        ksize=[1, size, size, 1],
        strides=[1, size, size, 1],
        padding='SAME'
    )


def model_pass(input, params, is_training):
    """
    Obavlja jedan prolaz kroz ceo model (forward pass).
    
    Parametri
    ----------
    input : Tensor
        Ulazni batch slika.
    params : Parameters
        Namedtuple sa parametrima modela.
    is_training : tf.bool Tensor
        Flag koji označava da li se radi treniranje (koristi se za dropout).
    
    Returns
    -------
    logits : Tensor
        Neaktivirane vrednosti na izlazu (pre softmax-a).
    """
    
    # Prvi konvolucioni sloj + ReLU
    with tf.variable_scope('conv1'):
        conv1 = conv_relu(input, kernel_size=params.conv1_k, depth=params.conv1_d)
    
    # Pooling sloj + dropout ako treniramo
    with tf.variable_scope('pool1'):
        pool1 = pool(conv1, size=2)
        pool1 = tf.cond(is_training, lambda: tf.nn.dropout(pool1, keep_prob=params.conv1_p), lambda: pool1)
    
    # Drugi konvolucioni sloj + ReLU
    with tf.variable_scope('conv2'):
        conv2 = conv_relu(pool1, kernel_size=params.conv2_k, depth=params.conv2_d)
    
    # Pooling + dropout
    with tf.variable_scope('pool2'):
        pool2 = pool(conv2, size=2)
        pool2 = tf.cond(is_training, lambda: tf.nn.dropout(pool2, keep_prob=params.conv2_p), lambda: pool2)
    
    # Treći konvolucioni sloj + ReLU
    with tf.variable_scope('conv3'):
        conv3 = conv_relu(pool2, kernel_size=params.conv3_k, depth=params.conv3_d)
    
    # Pooling + dropout
    with tf.variable_scope('pool3'):
        pool3 = pool(conv3, size=2)
        pool3 = tf.cond(is_training, lambda: tf.nn.dropout(pool3, keep_prob=params.conv3_p), lambda: pool3)
    
    # Dalje radimo dodatni pooling i flattenovanje na različitim nivoima izlaza
    
    # 1. nivo - dodatni pooling
    pool1 = pool(pool1, size=4)
    shape = pool1.get_shape().as_list()
    pool1 = tf.reshape(pool1, [-1, shape[1] * shape[2] * shape[3]])
    
    # 2. nivo - dodatni pooling
    pool2 = pool(pool2, size=2)
    shape = pool2.get_shape().as_list()
    pool2 = tf.reshape(pool2, [-1, shape[1] * shape[2] * shape[3]])
    
    # 3. nivo - samo flattenovanje
    shape = pool3.get_shape().as_list()
    pool3 = tf.reshape(pool3, [-1, shape[1] * shape[2] * shape[3]])
    
    # Spajanje svih feature mapa u jedan vektor
    flattened = tf.concat([pool1, pool2, pool3], axis=1)
    
    # Fully connected sloj sa ReLU i dropout-om
    with tf.variable_scope('fc4'):
        fc4 = fully_connected_relu(flattened, size=params.fc4_size)
        fc4 = tf.cond(is_training, lambda: tf.nn.dropout(fc4, keep_prob=params.fc4_p), lambda: fc4)
    
    # Izlazni sloj (logits)
    with tf.variable_scope('out'):
        logits = fully_connected(fc4, size=params.num_classes)
    
    return logits
```

```python
def plot_curve(axis, params, train_column, valid_column, linewidth=2, train_linestyle="b-", valid_linestyle="g-"):
    """
    Iscrtava grafikon trening i validacionih metrika (npr. accuracy ili loss) na zadatoj osi.

    Parameters
    ----------
    axis : matplotlib.axes.Axes
        Osa na kojoj će se crtati grafikon.
    params : Parameters
        Namedtuple sa parametrima modela (koristi se za pronalaženje fajla sa istorijom treninga).
    train_column : str
        Naziv kolone sa trening vrednostima u snimljenoj istoriji (npr. 'train_accuracy_history').
    valid_column : str
        Naziv kolone sa validacionim vrednostima (npr. 'valid_accuracy_history').
    linewidth : int, optional
        Debljina linije (default 2).
    train_linestyle : str, optional
        Stil linije za trening skup (default plava linija "b-").
    valid_linestyle : str, optional
        Stil linije za validacioni skup (default zelena linija "g-").

    Returns
    -------
    int
        Broj epoha prikazanih na grafiku.
    """
    # Učitavamo snimljenu istoriju treninga iz .npz fajla (train_history.npz)
    model_history = np.load(Paths(params).train_history_path + ".npz")

    # Uzima vrednosti trening i validacionih metrika
    train_values = model_history[train_column]
    valid_values = model_history[valid_column]

    # Broj epoha (dužina niza vrednosti)
    epochs = train_values.shape[0]

    # X osa (epohe od 0 do poslednje)
    x_axis = np.arange(epochs)

    # Crtamo samo one vrednosti koje su veće od nule (za svaki skup)
    axis.plot(x_axis[train_values > 0], train_values[train_values > 0], train_linestyle, linewidth=linewidth, label="train")
    axis.plot(x_axis[valid_values > 0], valid_values[valid_values > 0], valid_linestyle, linewidth=linewidth, label="valid")

    return epochs


def plot_learning_curves(params):
    """
    Iscrtava krive učenja (accuracy i loss za trening i validaciju) za model definisan parametrima.

    Parameters
    ----------
    params : Parameters
        Namedtuple sa parametrima modela.

    Returns
    -------
    None
    """
    # Kreira novu figuru sa 2 subplot-a u jednom redu
    curves_figure = pyplot.figure(figsize=(10, 4))

    # Prvi subplot za accuracy
    axis = curves_figure.add_subplot(1, 2, 1)
    epochs_plotted = plot_curve(axis, params, train_column="train_accuracy_history", valid_column="valid_accuracy_history")
    pyplot.grid()
    pyplot.legend()
    pyplot.xlabel("epoch")
    pyplot.ylabel("accuracy")
    pyplot.ylim(50., 115.)  # Y osa od 50% do 115%
    pyplot.xlim(0, epochs_plotted)

    # Drugi subplot za loss
    axis = curves_figure.add_subplot(1, 2, 2)
    epochs_plotted = plot_curve(axis, params, train_column="train_loss_history", valid_column="valid_loss_history")
    pyplot.grid()
    pyplot.legend()
    pyplot.xlabel("epoch")
    pyplot.ylabel("loss")
    pyplot.ylim(0.0001, 10.)  # Y osa od 0.0001 do 10 (log skala)
    pyplot.xlim(0, epochs_plotted)
    pyplot.yscale("log")  # Logaritamska skala na Y osi za loss

    # Prikazuje grafike (može se pozvati i spolja da se prikažu)
    pyplot.show()
```

```python
def train_model(params, X_train, y_train, X_valid, y_valid, X_test, y_test, logger_config):
    """
    Obavlja trening modela na osnovu datih trening podataka i parametara, 
    potom vrši evaluaciju na test skupu podataka.
    Deo trening skupa može biti korišćen kao validacioni tokom treninga ukoliko je zadato.

    Parameters
    ----------
    params : Parameters
        Namedtuple koji sadrži sve parametre modela.
    X_train : np.array
        Trening ulazni podaci.
    y_train : np.array
        Label-e za trening podatke.
    X_valid : np.array
        Validacioni ulazni podaci.
    y_valid : np.array
        Label-e za validacione podatke.
    X_test : np.array
        Test ulazni podaci.
    y_test : np.array
        Label-e za test podatke.
    logger_config : dict
        Konfiguracija loggera za Dropbox i Telegram obaveštenja.

    """

    # Priprema putanja i kreiranje loggera za zapis i obaveštenja
    paths = Paths(params)
    log = ModelCloudLog(
        os.path.join(paths.root_path, "logs"),
        dropbox_token=logger_config["dropbox_token"],
        telegram_token=logger_config["telegram_token"],
        telegram_chat_id=logger_config["telegram_chat_id"]
    )
    start = time.time()  # pamti vreme početka treninga
    model_variable_scope = paths.var_scope  # ime varijabilnog scope-a za model

    # Logujemo parametre treninga i veličine skupova podataka
    log.log_parameters(params, y_train.shape[0], y_valid.shape[0], y_test.shape[0]) 

    # Kreiramo TensorFlow graf
    graph = tf.Graph()
    with graph.as_default():
        # Definišemo placeholder-e za ulazne podatke i label-e (batches)
        tf_x_batch = tf.placeholder(tf.float32, shape=(None, params.image_size[0], params.image_size[1], 1))
        tf_y_batch = tf.placeholder(tf.float32, shape=(None, params.num_classes))
        is_training = tf.placeholder(tf.bool)  # indikator režima treninga ili evaluacije
        current_epoch = tf.Variable(0, trainable=False)  # broj epoha (za learning rate decay)

        # Podešavanje learning rate-a sa ili bez opadajuće stope
        if params.learning_rate_decay:
            learning_rate = tf.train.exponential_decay(
                params.learning_rate,
                current_epoch,
                decay_steps=params.max_epochs,
                decay_rate=0.01
            )
        else:
            learning_rate = params.learning_rate

        # Izgradnja modela unutar zadatog var_scope-a
        with tf.variable_scope(model_variable_scope):
            logits = model_pass(tf_x_batch, params, is_training)

            # L2 regularizacija samo ako je omogućena
            if params.l2_reg_enabled:
                with tf.variable_scope('fc4', reuse=True):
                    l2_loss = tf.nn.l2_loss(tf.get_variable('weights'))
            else:
                l2_loss = 0

        # Izračunavanje predikcija i gubitka
        predictions = tf.nn.softmax(logits)
        softmax_cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=tf_y_batch)
        loss = tf.reduce_mean(softmax_cross_entropy) + params.l2_lambda * l2_loss  

        # Definisanje optimizatora (Adam)
        optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)

    # Pokretanje TF sesije sa definisanim grafom
    with tf.Session(graph=graph) as session:
        session.run(tf.global_variables_initializer())  # inicijalizacija svih varijabli

        # Funkcija za evaluaciju modela po batch-evima (računa accuracy i loss)
        def get_accuracy_and_loss_in_batches(X, y):
            p = []
            sce = []
            batch_iterator = BatchIterator(batch_size=128)
            for x_batch, y_batch in batch_iterator(X, y):
                [p_batch, sce_batch] = session.run([predictions, softmax_cross_entropy], feed_dict={
                    tf_x_batch: x_batch,
                    tf_y_batch: y_batch,
                    is_training: False
                })
                p.extend(p_batch)
                sce.extend(sce_batch)
            p = np.array(p)
            sce = np.array(sce)
            accuracy = 100.0 * np.sum(np.argmax(p, 1) == np.argmax(y, 1)) / p.shape[0]
            loss = np.mean(sce)
            return accuracy, loss

        # Ukoliko je podešeno da nastavimo trening od ranije, pokušavamo da učitamo model
        if params.resume_training:
            try:
                tf.train.Saver().restore(session, paths.model_path)
            except Exception as e:
                log("Failed restoring previously trained model: file does not exist.")
                pass

        saver = tf.train.Saver()
        early_stopping = EarlyStopping(tf.train.Saver(), session, patience=params.early_stopping_patience, minimize=True)

        # Liste za beleženje istorije gubitka i tačnosti
        train_loss_history = np.empty([0], dtype=np.float32)
        train_accuracy_history = np.empty([0], dtype=np.float32)
        valid_loss_history = np.empty([0], dtype=np.float32)
        valid_accuracy_history = np.empty([0], dtype=np.float32)

        # Ako je broj epoha pozitivan, krećemo sa treningom, inače samo testiramo
        if params.max_epochs > 0:
            log("================= TRAINING ==================")
        else:
            log("================== TESTING ==================")
        log(" Timestamp: " + get_time_hhmmss())
        log.sync()

        # Glavna petlja treninga po epohama
        for epoch in range(params.max_epochs):
            current_epoch = epoch

            # Treniranje na celom skupu u batch-evima
            batch_iterator = BatchIterator(batch_size=params.batch_size, shuffle=True)
            for x_batch, y_batch in batch_iterator(X_train, y_train):
                session.run([optimizer], feed_dict={
                    tf_x_batch: x_batch,
                    tf_y_batch: y_batch,
                    is_training: True
                })

            # Logovanje metrika na zadatim intervalima
            if (epoch % params.log_epoch == 0):
                valid_accuracy, valid_loss = get_accuracy_and_loss_in_batches(X_valid, y_valid)
                train_accuracy, train_loss = get_accuracy_and_loss_in_batches(X_train, y_train)

                if (epoch % params.print_epoch == 0):
                    log("-------------- EPOCH %4d/%d --------------" % (epoch, params.max_epochs))
                    log("     Train loss: %.8f, accuracy: %.2f%%" % (train_loss, train_accuracy))
                    log("Validation loss: %.8f, accuracy: %.2f%%" % (valid_loss, valid_accuracy))
                    log("      Best loss: %.8f at epoch %d" % (early_stopping.best_monitored_value, early_stopping.best_monitored_epoch))
                    log("   Elapsed time: " + get_time_hhmmss(start))
                    log("      Timestamp: " + get_time_hhmmss())
                    log.sync()
            else:
                # Za epohe bez logovanja postavljamo vrednosti na 0 da ih ne bi beležili
                valid_loss = 0.
                valid_accuracy = 0.
                train_loss = 0.
                train_accuracy = 0.

            # Čuvanje istorije metrika
            valid_loss_history = np.append(valid_loss_history, [valid_loss])
            valid_accuracy_history = np.append(valid_accuracy_history, [valid_accuracy])
            train_loss_history = np.append(train_loss_history, [train_loss])
            train_accuracy_history = np.append(train_accuracy_history, [train_accuracy])

            # Provera ranog zaustavljanja ako je omogućeno
            if params.early_stopping_enabled:
                # Ako valid_loss nije izračunat, izračunavamo ga sada
                if valid_loss == 0:
                    _, valid_loss = get_accuracy_and_loss_in_batches(X_valid, y_valid)
                # Provera da li je vreme za rano zaustavljanje i vraćanje najboljih parametara
                if early_stopping(valid_loss, epoch):
                    log("Early stopping.\nBest monitored loss was {:.8f} at epoch {}.".format(
                        early_stopping.best_monitored_value, early_stopping.best_monitored_epoch
                    ))
                    break

        # Evaluacija na test skupu
        test_accuracy, test_loss = get_accuracy_and_loss_in_batches(X_test, y_test)
        valid_accuracy, valid_loss = get_accuracy_and_loss_in_batches(X_valid, y_valid)

        # Logovanje završnih rezultata i vremena
        log("=============================================")
        log(" Valid loss: %.8f, accuracy = %.2f%%)" % (valid_loss, valid_accuracy)) 
        log(" Test loss: %.8f, accuracy = %.2f%%)" % (test_loss, test_accuracy)) 
        log(" Total time: " + get_time_hhmmss(start))
        log("  Timestamp: " + get_time_hhmmss())

        # Čuvanje modela i istorije treninga na disk
        saved_model_path = saver.save(session, paths.model_path)
        log("Model file: " + saved_model_path)
        np.savez(paths.train_history_path,
                 train_loss_history=train_loss_history,
                 train_accuracy_history=train_accuracy_history,
                 valid_loss_history=valid_loss_history,
                 valid_accuracy_history=valid_accuracy_history)
        log("Train history file: " + paths.train_history_path)

        # Sinhronizacija logova i obaveštenja o završetku treninga
        log.sync(notify=True, message="Finished training with *%.2f%%* accuracy on the testing set (loss = *%.6f*)." % (test_accuracy, test_loss))

        # Prikazivanje grafika učenja i slanje na cloud
        plot_learning_curves(params)
        log.add_plot(notify=True, caption="Learning curves")

        pyplot.show()

```

```python
import os
import pickle
import numpy as np
from skimage.transform import resize


def preprocess_and_save_chunks(src_dir, dst_dir, target_size=(32, 32, 3)):
    """
    Učitaj svaki chunk iz src_dir, resize slike i sačuvaj preprocesirani chunk u dst_dir.
    """
    os.makedirs(dst_dir, exist_ok=True)

    files = sorted([f for f in os.listdir(src_dir) if f.endswith('.p')])

    for fname in files:
        src_path = os.path.join(src_dir, fname)
        dst_path = os.path.join(dst_dir, fname)

        print(f"Processing chunk {fname} ...")

        with open(src_path, 'rb') as f:
            data = pickle.load(f)

        X = data["features"]
        y = data["labels"]

        # Resize slike u chunk-u
        X_processed = np.array([resize(img, target_size, anti_aliasing=True) for img in X], dtype=np.float32)

        # Sačuvaj preprocesirani chunk
        with open(dst_path, 'wb') as f:
            pickle.dump({"features": X_processed, "labels": y}, f)

        print(f"Saved preprocessed chunk to {dst_path}")


def chunk_generator(chunks_dir):
    """
    Generator koji učitava jedan preprocesirani chunk u memoriju po pozivu.

    Args:
        chunks_dir (str): Folder sa preprocesiranim chunkovima.

    Yields:
        Tuple[np.ndarray, np.ndarray]: features i labels za jedan chunk.
    """
    files = sorted([f for f in os.listdir(chunks_dir) if f.endswith('.p')])

    for fname in files:
        path = os.path.join(chunks_dir, fname)
        with open(path, 'rb') as f:
            data = pickle.load(f)
            yield data["features"], data["labels"]


# --- Putanje ---
train_balanced_chunks_dir = "traffic-signs-data/balanced_chunks"
train_balanced_preprocessed_chunks_dir = "traffic-signs-data/balanced_chunks_preprocessed"

train_extended_chunks_dir = "traffic-signs-data/extended_chunks"
train_extended_preprocessed_chunks_dir = "traffic-signs-data/extended_chunks_preprocessed"

test_chunks_dir = "traffic-signs-data/test_chunks"
test_preprocessed_chunks_dir = "traffic-signs-data/test_chunks_preprocessed"

# --- Preprocess i sačuvaj po chunku ---
preprocess_and_save_chunks(train_balanced_chunks_dir, train_balanced_preprocessed_chunks_dir)
preprocess_and_save_chunks(train_extended_chunks_dir, train_extended_preprocessed_chunks_dir)
#preprocess_and_save_chunks(test_chunks_dir, test_preprocessed_chunks_dir)

# --- Upotreba generatora za obradu po chunku ---
print("\nIterating through balanced preprocessed chunks:")
for X_chunk, y_chunk in chunk_generator(train_balanced_preprocessed_chunks_dir):
    print(f"Chunk shape: {X_chunk.shape}, Labels shape: {y_chunk.shape}")
    # Example: ovde bi išla obuka modela nad ovim chunk-om
    # model.train_on_batch(X_chunk, y_chunk)

print("\nIterating through extended preprocessed chunks:")
for X_chunk, y_chunk in chunk_generator(train_extended_preprocessed_chunks_dir):
    print(f"Chunk shape: {X_chunk.shape}, Labels shape: {y_chunk.shape}")
```

```python
def get_top_k_predictions(params, X, k = 5):
    """
    Evaluira skup podataka `X` koristeći model definiran parametrima `params`
    i vraća top k predikcija (po zadanoj vrijednosti k, default je 5).

    Parametri
    ----------
    params    : Parameters
                Struktura (`namedtuple`) koja sadrži parametre modela.
    X         : 
                Skup za testiranje (npr. slike).
    k         : 
                Broj top predikcija koje želimo dobiti.

    Povratna vrijednost
    -------------------
    Numpy niz s top k softmax predikcija za svaki primjer iz `X`.
    """
    
    # Inicijalizacija varijabli i putanja
    paths = Paths(params)
    
    # Kreiranje TensorFlow grafa
    graph = tf.Graph()
    with graph.as_default():
        # Placeholder za ulazne podatke sa dimenzijama (batch_size, visina, širina, kanali)
        tf_x = tf.placeholder(tf.float32, shape = (None, params.image_size[0], params.image_size[1], 1))
        # Oznaka da model nije u fazi treniranja (is_training = False)
        is_training = tf.constant(False)
        
        # Kreiranje varijabilnog prostora (scope) za model prema var_scope
        with tf.variable_scope(paths.var_scope):
            # Dobijanje predikcija modela (softmax vektor vjerovatnoća)
            predictions = tf.nn.softmax(model_pass(tf_x, params, is_training))
            # Dohvat top k vjerovatnoća i njihovih indeksa
            top_k_predictions = tf.nn.top_k(predictions, k)

    # Pokretanje sesije za izvođenje grafa
    with tf.Session(graph = graph) as session:
        session.run(tf.global_variables_initializer())
        # Učitavanje sačuvanih težina modela
        tf.train.Saver().restore(session, paths.model_path)
        # Izvršavanje grafa za top k predikcija sa ulazom X
        [p] = session.run([top_k_predictions], feed_dict = {tf_x : X})
        
        # Vraćanje top k predikcija kao numpy niz
        return np.array(p)
```
